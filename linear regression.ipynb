```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("C:\\Users\\sgupta1\\Desktop\\Realworld Projects\\Project -1\\tvmarketing.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.pairplot(df, x_vars=['TV'], y_vars='Sales',size=7, aspect=0.7, kind='scatter')
```

```python
X = df['TV']
y = df['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9 , random_state=100)
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
```

```python
X_train = X_train[:, np.newaxis]
X_test = X_test[:, np.newaxis]
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_test.shape
```

```python
y_pred.shape
```

```python
c = [i for i in range(1,21,1)]         # generating index 
fig = plt.figure()
plt.plot(c,y_test, color="blue", linewidth=2.5, linestyle="-")
plt.plot(c,y_pred, color="red",  linewidth=2.5, linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                               # X-label
plt.ylabel('Sales', fontsize=16) 
plt.show()
```

```python
c = [i for i in range(1,21,1)]
fig = plt.figure()
plt.plot(c,y_test-y_pred, color="blue", linewidth=2.5, linestyle="-")
fig.suptitle('Error Terms', fontsize=20)              # Plot heading 
plt.xlabel('Index', fontsize=18)                      # X-label
plt.ylabel('ytest-ypred', fontsize=16)   
```
